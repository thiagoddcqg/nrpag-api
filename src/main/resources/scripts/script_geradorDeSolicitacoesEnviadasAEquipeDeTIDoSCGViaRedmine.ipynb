```python
#ESCRITA - tarefasSCG.csv
with open(pathSCG, 'w', newline='', encoding='utf-8') as file2:

    writer = csv.writer(file2)
    
    writer.writerow(["#",                                 #1
                     "Sistema-Modulo",                    #2
                     "Projeto-Modulo",                    #3
                     "# Projeto Hierarquia",              #4
                     "Projeto Hierarquia",                #5
                     "Tipo",                              #6
                     "Situação",                          #7
                     "Título",                            #8
                     "Atribuído para",                    #9
                     "Catálogo_01 (Memora - 13/2020)",    #10
                     "HET (Real)",                        #11
                     "qtd_servicos",                      #12
                     "Início",                            #13
#                      "Concluído",                         #14
                     "Tarefa pai",                        #15
                     "Qtd. de Anexos"])                   #16

    i=0
    for tarefaSCG in tarefasSCG:

        i = i+1

        if hasattr(tarefaSCG, "parent"):
            try:
                id_projeto = redmine.issue.get(tarefaSCG.parent).parent
                nome_projeto = redmine.issue.get(id_projeto).project
            except:
                id_projeto = tarefaSCG.parent
                nome_projeto = redmine.issue.get(id_projeto).project

            tarefaPai = tarefaSCG.parent
        else:
            id_projeto = ""
            nome_projeto = ""
            tarefaPai = ""

        if hasattr(tarefaSCG, "spent_hours"):
            horasGastas = tarefaSCG.spent_hours
        else:
            horasGastas = ""

        if hasattr(tarefaSCG, "assigned_to"):
            atribuidoPara = tarefaSCG.assigned_to
        else:
            atribuidoPara = ""

        if hasattr(tarefaSCG, "start_date"):
            dataInicio = tarefaSCG.start_date
        else:
            dataInicio = ""

        sistemamodulo     = ""
        catalogoCast      = ""
        catalogoMemoraI   = ""
        catalogoMemoraII  = ""
        qtdServicosMemora = ""

        for cf in tarefaSCG['custom_fields']:
            if cf['name'] == "Catálogo (Cast - 09/2020)":
                catalogoCast = cf['value'].replace(" ", "")
            elif cf['name'] == "Catálogo (Memora - 42/2018)":
                catalogoMemoraI = cf['value'].replace(" ", "")
            elif cf['name'] == "Catálogo_01 (Memora - 13/2020)":
                catalogoMemoraII = cf['value'].replace(" ", "")
            elif cf['name'] == "Sistema (Módulo)":
                sistemamodulo = cf['value']
            elif cf['name'] == "Qtd Serviços (Memora-13/2020)":
                qtdServicosMemora = cf['value']

        print("Processando " + str(i) + " de " + str(len(tarefasSCG)))

        writer.writerow([str(tarefaSCG.id),                           #1
                         str(sistemamodulo),                          #2
                         str(tarefaSCG.project),                      #3
                         str(id_projeto),                             #4
                         str(nome_projeto),                           #5
                         str(tarefaSCG.tracker),                      #6
                         str(tarefaSCG.status),                       #7
                         str(tarefaSCG.subject),                      #8
                         str(atribuidoPara),                          #9
                         str(catalogoMemoraII),                       #10
                         str(horasGastas),                            #11
                         str(qtdServicosMemora),                      #12
                         str(dataInicio),                             #13
#                          str(tarefaSCG.closed_on),                    #14
                         str(tarefaPai),                              #15
                         str(len(tarefaSCG.attachments))])            #16
    file2.close()
```
